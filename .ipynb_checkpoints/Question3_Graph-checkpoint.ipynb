```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
exp = ['1e3', '1e4', '1e5', '1e6', '1e7']
heap_exp3 = np.mean(np.array([0.014, 0.014]))
count_exp3 = np.mean(np.array([0.012, 0.011]))
radix_exp3 = np.mean(np.array([0.012, 0.011]))

heap_exp4 = np.mean(np.array([0.178, 0.189]))
count_exp4 = np.mean(np.array([0.167, 0.163]))
radix_exp4 = np.mean(np.array([0.192, 0.205]))

heap_exp5 = np.mean(np.array([0.376, 0.442]))
count_exp5 = np.mean(np.array([0.318, 0.331]))
radix_exp5 = np.mean(np.array([0.327, 0.351]))

heap_exp6 = np.mean(np.array([2.141, 2.629]))
count_exp6 = np.mean(np.array([1.575, 1.689]))
radix_exp6 = np.mean(np.array([1.734, 1.781]))

heap_exp7 = np.mean(np.array([19.33]))
count_exp7 = np.mean(np.array([12.97]))
radix_exp7 = np.mean(np.array([13.96]))

heap_ = [heap_exp3, heap_exp4, heap_exp5, heap_exp6, heap_exp7]
count_ = [count_exp3, count_exp4, count_exp5, count_exp6, count_exp7]
radix_ = [radix_exp3, radix_exp4, radix_exp5, radix_exp6, radix_exp7]
```

```python
plt.plot(exp, heap_, label='heap sort')
plt.plot(exp, count_, label='count sort')
plt.plot(exp, radix_, label='radix sort')
plt.xlabel('Size of input')
plt.ylabel('Time taken')

plt.legend()
```
